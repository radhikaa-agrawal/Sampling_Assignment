```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Creditcard_data.csv")
df.head()
```

```python
from sklearn.preprocessing import normalize
Amount = normalize([df['Amount']])[0]
df['Amount'] = Amount
df = df.iloc[:, 1:]
df.head()
```

```python
class_frequency = df['Class'].value_counts()
print(f"Number of classes: {len(class_frequency)}")
print("Class frequencies:")
print(class_frequency)
```

```python
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

X = df.drop('Class', axis=1)
y = df['Class']

oversampler = RandomOverSampler()
undersampler = RandomUnderSampler()

X_over, y_over = oversampler.fit_resample(X, y)
X_resampled, y_resampled = undersampler.fit_resample(X_over, y_over)

print("Total number of samples in each class after resampling:")
print(y_resampled.value_counts())
```

```python
X_resampled
```

```python
df_new = pd.concat([X_resampled, y_resampled], axis=1)
df_new
```

```python
class_frequency = df_new['Class'].value_counts()
print(f"Number of classes: {len(class_frequency)}")
print("Class frequencies:")
print(class_frequency)
```

```python
# Simple Random Sampling
n= int((1.96*1.96 * 0.5*0.5)/(0.05**2))
sampled_df = df_new.sample(n=n, random_state=42)
print(sampled_df)
```

```python
# Systematic Sampling

Interval = 2
systematic_df = df_new.iloc[::Interval]
systematic_df
```

```python
# Stratified Sampling

from sklearn.model_selection import train_test_split

n = int((1.96*1.96 * 0.5*0.5)/((0.05)**2))

Strata = df_new.groupby('Class')

stratified_df = Strata.apply(lambda x: x.sample(n))
stratified_df
```

```python
#Cluster Sampling
import numpy as np

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10, random_state=42).fit(df_new)
clusterLabels = kmeans.labels_

selected_clusters = [0, 2, 4, 5, 8]

clusterSeries = pd.Series(clusterLabels)

df_clusterSample = df_new[clusterSeries.isin(selected_clusters)]

(df_clusterSample)
```

```python
# Convenience sampling
convenience_sample = pd.concat([df_new.head(380), df_new.tail(380)])
convenience_sample
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# Spliting the data into training and testing sets
train_dfs = []
test_dfs = []
combinedsampledf=[systematic_df,sampled_df,stratified_df,convenience_sample,df_clusterSample]
for dff in combinedsampledf:
    X = dff.drop('Class', axis=1)
    y = dff['Class']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    train_dfs.append(pd.concat([X_train, y_train], axis=1))
    test_dfs.append(pd.concat([X_test, y_test], axis=1))

# Initializing the models
DecisionTree = DecisionTreeClassifier()
KNN = KNeighborsClassifier()
GaussiaNB = GaussianNB()
Rf = RandomForestClassifier()
Svm = SVC()

# Training and testing each model
results = []
modelName=["Decision Tree Classifier","KNN Classifier","Gaussian Naive Bayes","Random Forest","Support Vector Machine"]
samplerName=["Simple Random Sampling","Systematic Sampling","Stratified Sampling","Cluster Sampling","Convenience Sampling"]
for i in range(len(combinedsampledf)):
    train_df = train_dfs[i]
    test_df = test_dfs[i]
    model_results = []
    x=0
    for model in [DecisionTree, KNN, GaussiaNB, Rf, Svm]:
        #model_name = type(model).name
        model.fit(train_df.drop('Class', axis=1), train_df['Class'])
        accuracy = model.score(test_df.drop('Class', axis=1), test_df['Class'])
        model_results.append([modelName[x],accuracy])
        x+=1
    results.append(pd.DataFrame(model_results, columns=['Model','Accuracy']))

# Concatenating the results into a single DataFrame

final_results = pd.concat(results, keys=[i for i in samplerName])
```
